### BAR CHART

```python

# Plotting the bar chart
plt.figure(figsize=(10, 6))  # Set the figure size
plt.bar(df['Country/Territory'], df['Capital'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Capital')
plt.title('Population Distribution by Country')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Sample 20% of the data
sampled_df = df.sample(frac=0.2, random_state=42)

# Display the sampled data to verify
print(sampled_df.head())
```

```python

# Plotting the bar chart
plt.figure(figsize=(10, 6))  # Set the figure size
plt.bar(sampled_df['Country/Territory'], sampled_df['Capital'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Capital')
plt.title('Population Distribution by Country')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

### HISTOGRAM

```python
plt.figure(figsize=(20, 12))
plt.hist(sampled_df['2022 Population'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Population')
plt.ylabel('World Population Percentage')
plt.grid(True)
plt.show()
```
